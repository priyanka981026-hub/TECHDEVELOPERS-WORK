# Python Seaborn 

```python
 #  Python Seaborn --is a library mostly used for statistical plotting in python.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load movie metadata
df = pd.read_csv("movies_metadata.csv", low_memory=False)

# Convert vote_average to numeric
df['vote_average'] = pd.to_numeric(df['vote_average'], errors='coerce')
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

# Drop NaN for relevant plotting
df = df[['title', 'budget', 'revenue', 'vote_average']].dropna().head(1000)  # sample
```

```python
# 1. Using Seaborn + Matplotlib (title + limits)
sns.lineplot(x='budget', y='revenue', data=df)
plt.title("Budget vs Revenue")
plt.xlim(0, 100000000) # zoom in
plt.show()
```

```python
# 2. Set Style (darkgrid, white, etc.)
sns.set_style("whitegrid")
sns.scatterplot(x='budget',y='vote_average', data=df)
plt.show()
```

```python
# 3. Despine
sns.boxplot(x='vote_average', data=df)
sns.despine()
plt.show()
```

```python
# 4. Change Figure Size
plt.figure(figsize=(8,4))
sns.histplot(df['vote_average'])
plt.show()
```

```python
# 5. Scaling Plots (paper, talk, poster)
sns.set_context("talk", font_scale=1)
sns.kdeplot(df['vote_average'], shade=True)
plt.show()
```

```python
 # 6. Temporarily Set Style
def plot():
    sns.violinplot(x='vote_average', data=df)
with sns.axes_style('dark'):
    plt.subplot(211)
    plot()
plt.subplot(212)
plot()
plt.show()
```

```python
# 7. Color Palettes
sns.set_palette("Accent")
sns.histplot(df['vote_average'])
plt.show()
```

```python
# Sequential:
palette = sns.color_palette("Blues",10)
sns.palplot(palette)
plt.show()
```

```python
# 8. Multiple Plots (subplot)
def graph():
    sns.scatterplot(x='budget', y='revenue', data=df)
plt.subplot(121)
graph()
plt.subplot(122)
graph()
plt.show()
```

```python
# 9. FacetGrid
# Add a fake 'category' for demonstration
df['group'] = pd.cut(df['vote_average'], bins=[0,5,7,10], labels=["Low","mid","high"])
g = sns.FacetGrid(df, col="group")
g.map_dataframe(sns.histplot, x="vote_average")
plt.show()
```

```python
# Lambda with if-else
# check if a number  is even or odd
check = lambda x: "Even" if x % 2 == 0 else "odd"
print(check(4))
print(check(7))
```

```python
 # Lambda with Multiple Statements
# Example: Perform addition and multiplication in a single line
calc = lambda x,y:(x+y,x*y)
res = calc(3,4)
print(res)
```

```python
# 1. Lambda with Condition Checking
df['rating_tag'] = df['vote_average'].apply(lambda x: "high" if x >=7 else "medium" if x >=5 else "low")
print("\n----Rating Tags ---")
print(df[['title','vote_average','rating_tag']].dropna().head(5))
```

```python
# 2. Lambda with filter()
high_rated = list(
    filter(lambda x: x >=7, df['vote_average'].dropna())
)
print("high rated movies (first 5):", high_rated[:5])
```

```python
# 3. Lambda with map()
# Increase vote_average by 10%.
adjusted_votes = list(
    map(lambda x: round(x * 1.1,1), df['vote_average'].dropna())
)
print("Adjusted votes (first 5):",adjusted_votes[:5])
```

```python
 # 4. Lambda with reduce()
from functools import reduce
vote_sum = reduce(
    lambda x,y: x+y, df['vote_average'].dropna().head(5)
)
print("sum of top 5 vote_averages:", vote_sum)
```

```python
# 5. Lambda with List Comprehension
# Convert top 5 movie titles to uppercase.
titles = df['title'].dropna().head(5)
upper_titles = [(lambda t=title: t.upper())() for title in titles]
print("Top  5 titles  in uppercase:",upper_titles)
```
