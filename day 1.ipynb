```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Required for 3D projection

x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')  # ✅ use '3d' not '3D'
ax.plot(x, y, z)
plt.title("3D Sine Wave")
plt.show()
```

```python
# 2. Visualize Bubble Sort
# import matplotlib.pyplot as plt
# import time
arr=[8,4,2,6,1]
for i in range (len(arr)):
    for j in range(len(arr)-i-1):
        if arr[j]>arr[j+1]:
            arr[j],arr[j+1]=arr[j+1],arr[j]
            plt.clf()
            plt.bar(range(len(arr)),arr,color='skyblue')
            plt.title("Bubble sort")
            plt.pause(0.3)
plt.show()            
```

```python
# 3. Merge Sort Visualization
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i = j = k = 0

        while i < len(L) and j < len(R):
            arr[k] = L[i] if L[i] < R[j] else R[j]
            i += L[i] < R[j]
            j += L[i] >= R[j]
            k += 1
            plt.clf()
            plt.bar(range(len(arr)), arr)
            plt.title("Merge Sort")
            plt.pause(0.2)

        while i < len(L): arr[k], i, k = L[i], i+1, k+1
        while j < len(R): arr[k], j, k = R[j], j+1, k+1

arr = [7, 3, 8, 5, 2]
merge_sort(arr)
plt.show()
```

```python
# 4. Quick Sort Visualization
def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi)
        quick_sort(arr, pi+1, high)

def partition(arr, low, high):
    pivot = arr[low]
    left = low + 1
    right = high
    while True:
        while left <= right and arr[left] <= pivot: left += 1
        while arr[right] > pivot: right -= 1
        if left > right: break
        arr[left], arr[right] = arr[right], arr[left]
    arr[low], arr[right] = arr[right], arr[low]
    plt.clf()
    plt.bar(range(len(arr)), arr, color='green')
    plt.title("Quick Sort")
    plt.pause(0.3)
    return right

arr = [9, 2, 7, 5, 1]
quick_sort(arr, 0, len(arr)-1)
plt.show()
```

```python
# 5. 3D Quick Sort (Bar View)
from mpl_toolkits.mplot3d import Axes3D

arr = [5, 1, 7, 3, 9]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i in range(len(arr)):
    ax.clear()
    ax.bar3d(np.arange(len(arr)), np.zeros(len(arr)), np.zeros(len(arr)),
             dx=1, dy=1, dz=arr, color='cyan')
    arr.sort()
    plt.pause(0.5)
plt.show()
```

```python
 # 7. 3D Insertion Sort Visualization
arr = [4, 2, 9, 1, 6]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(1, len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and arr[j] > key:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key
    ax.clear()
    ax.bar3d(np.arange(len(arr)), np.zeros(len(arr)), np.zeros(len(arr)),
             dx=1, dy=1, dz=arr, color='red')
    plt.pause(0.5)

plt.show()
```

```python
 # 8. Time Series with Rolling Average (from Kaggle Movie Dataset)
import pandas as pd

movies = pd.read_csv('movies_metadata.csv', low_memory=False)
movies['release_date'] = pd.to_datetime(movies['release_date'], errors='coerce')
movies['vote_average'] = pd.to_numeric(movies['vote_average'], errors='coerce')
movies.set_index('release_date', inplace=True)

ts = movies['vote_average'].resample('Y').mean()
ts_rolling = ts.rolling(3).mean()

plt.plot(ts, label='Original')
plt.plot(ts_rolling, label='Rolling Avg')
plt.title("Time Series: Vote Avg with Rolling Mean")
plt.legend()
plt.show()
```
