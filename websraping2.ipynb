```python
import requests
from bs4 import BeautifulSoup   # bs4 --used to call beautifulsoup.
```

```python
web = requests.get("https://www.rbmi.in/") # response 200 means i can read and work on this websites
print(web)
```

```python
 web.content
```

```python
soup = BeautifulSoup(web.content,"html.parser")             # soup object call the class of beautifulsoup
```

```python
 lines = soup.find_all("p")
 lines
```

```python
 for l in lines:
     print(l.text)
```

```python
s = soup.find("div", class_="first-col-left")
                
```

```python
lines_1 = s.find_all("a") # anker tag
lines_1
```

```python
for l1 in lines_1:
    print(l1.text)
```
