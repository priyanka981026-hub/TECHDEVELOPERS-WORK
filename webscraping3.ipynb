# Extracting links in web page using python

```python
import requests
from bs4 import BeautifulSoup   # bs4 --used to call beautifulsoup.
```

```python
web = requests.get("https://www.rbmi.in/") # response 200 means i can read and work on this websites
print(web)
```

```python
# web.content
```

```python
soup = BeautifulSoup(web.content,"html.parser")             # soup object call the class of beautifulsoup
```

```python
#print(soup.prettify())
```

```python
soup.find_all("a")
```

```python
for i in soup.find_all("a"):
    print(i)
```

```python
for i in soup.find_all("a"):
    print(i.get("href"))
```

# Extracting image information in web page using python

```python
img = soup.find_all("img")
img
```

```python
for i in img:
    print(i.get("src"))
```

```python
for i in img:
    print(type(i.get("src")))
```
