```python
 #1. Linear Algebra in NumPy
# Matrix Multiplication & Manipulation 
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print("matrix multiplication:\n", np.dot(A,B))  # or A @ B
```

```python
# Matrix & Vector Products
v = np.array([1,2])
w = np.array([3,4])    # 1*3 + 2*4 = 11
print("dot product:",np.dot(v,w))
```

```python
# Determinants and Inverse
M = np.array([[2,1],[5,3]])
print("determinants:",np.linalg.det(M))
print("inverse:",np.linalg.inv(M))


```

```python
 # Inner and Outer Product
print("inner:",np.inner(v,w))  # inner product:dot product of two vectors(scalar result)
print("outerr:",np.outer(v,w))  #outer product:product of two vectors(matrix result)
```

```python
 # Dot and Vdot Functions
print("dot:",np.dot(v,w))  # np.dot():Dot product (matrix multiplication).
print("vdot:",np.vdot(v,w)) # np.vdot(): Dot product of vectors(with complex conjugate).
```

```python
# Eigenvalues and Eigenvectors
x = np.array([[2,-1],[-1,2]])  
eigenvalues,eigenvectors = np.linalg.eig(x) # eigenvalues:scalar values representing how much change occurs in a linear transformation.
print("eigenvalues:",eigenvalues)
print("eigenvectors:",eigenvectors) #eigenvectors:direction vectors that do not change direction under a linear transformation.
```

```python
# 2. Random Number Generation & Distributions
# Set seed for reproducibility
np.random.seed(42)
print("normal:",np.random.normal(0,1,5))  # mean=0, std=1
print("binomial:",np.random.binomial(10,0.5,5))  # 10 trials, p=0.5
print("poisson:",np.random.poisson(3,5))  # lambda=3
print("uniform:",np.random.uniform(1,10,5))  # low=1, high=10
print("exponential:",np.random.exponential(1,5))  # scale=1
print("chi square:",np.random.chisquare(2,5)) # df=2
```

```python
# 3. Statistical Functions (on movie data)
print("ratings - Mean:",np.mean(ratings))
print("ratings - Median:",np.median(ratings))
print("ratings - variance:",np.var(ratings))
print("ratings - std Dev:",np.std(ratings))
```

```python
# 4. Advanced NumPy Operations
# Vectorized Weighted Score (for movie ranking)
# Normalize vote_count

# Load dataset
df = pd.read_csv("movies_metadata.csv", low_memory=False)

# Select and clean data
df = df[['title', 'vote_average', 'vote_count']].dropna()
df['vote_count'] = pd.to_numeric(df['vote_count'], errors='coerce')
df = df.dropna()

# Convert to NumPy arrays (these will match the cleaned df)
ratings = df['vote_average'].to_numpy()
votes = df['vote_count'].to_numpy()

# 1. Normalize vote counts
norm_votes = (votes - votes.min()) / (votes.max() - votes.min())

# 2. Vectorized weighted score = rating × normalized vote count
df['score'] = ratings * norm_votes

# 3. Top 5 movies by score
top5 = df.sort_values(by='score', ascending=False)[['title', 'vote_average', 'vote_count', 'score']].head()

# 4. Print results
print(top5)
```
