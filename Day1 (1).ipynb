```python
#Pandas basics
import pandas as pd
 #sample data
```

```python
data = pd.read_csv("movies_metadata.csv")  
print(data.head())
```

## Day 2

```python
df = pd.DataFrame(data)
print(df)
```

```python
# PANDAS DATAFRAME INDEX


df = pd.DataFrame(data)
print(df. index)
```

```python
#2. Pandas DataFrame Index
# Show default index
print(df.index)

# Set a column as index (optional)
df_indexed = df.set_index('title')
print(df_indexed.head())
```

```python
# Null value check
df.isnull().sum().sort_values(ascending=False)
```

```python
# Convert to datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Convert budget and revenue to numeric
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

# Fill null runtimes with median
df['runtime'] = df['runtime'].fillna(df['runtime'].median())

# Drop rows with missing titles or release dates
df.dropna(subset=['title', 'release_date'], inplace=True)

print("After Cleaning Shape:", df.shape)
```

```python
# Top 5 release years (recent)
df['release_date'].dt.year.value_counts().sort_index(ascending=False).head()
```

```python
# Only English language movies
english_movies = df[df['original_language'] == 'en']
print("English Movies Count:", len(english_movies))
```

```python
# Sort by runtime
df[['title', 'runtime']].sort_values(by='runtime', ascending=False).head()
```

```python
# Filter by budget
high_budget = df[df['budget'] > 100_000_000]
high_budget[['title', 'budget']].sort_values(by='budget', ascending=False).head()
```

```python
#3. Access Data in DataFrame
#Access a single column
print(df['title'].head())

# Access multiple columns
print(df[['title', 'release_date']].head())

# Access by row/column position
print(df.iloc[0])           # first row
print(df.iloc[0, 1])        # first row, second column
print(df.loc[0, 'title'])   # row 0, column 'title'
```

```python
#4. Indexing and Selecting Data with Pandas
# Select rows based on a condition
df_recent = df[pd.to_datetime(df['release_date'], errors='coerce').dt.year > 2010]
print(df_recent[['title', 'release_date']].head())
```

```python
#5. Slicing Pandas DataFrame
# Slice first 5 rows
print(df[:5])

# Slice from row 10 to 20
print(df[10:21])
```

```python


#6 filter dataframe with multiple conditions
# Filter: budget > 10M and language is English
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df_filtered = df[(df['budget'] > 10000000) & (df['original_language'] == 'en')]
print(df_filtered[['title', 'budget', 'original_language']].head())
```

```python
#7 merging ,joining,concatenating, dataframes
df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['A', 'B']})
df2 = pd.DataFrame({'ID': [3, 4], 'Name': ['C', 'D']})

df_concat = pd.concat([df1, df2])
print(df_concat)
```

```python
#8. Sorting Pandas DataFrame
# Sort by runtime descending
df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')
df_sorted = df.sort_values(by='runtime', ascending=False)
print(df_sorted[['title', 'runtime']].head())
```

```python
#9 Pivot Table in Pandas
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['year'] = df['release_date'].dt.year
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')

# Pivot table
pivot = df.pivot_table(values='revenue', index='year', aggfunc='mean')
print(pivot.tail())

```
